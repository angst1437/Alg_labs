# Часть 1

создание структуры стека

```python
class Node:
    def __init__(self, data):
        self.next = None
        self.data = data

    
class Stack:
    def __init__(self):
        self.head = None

    def push(self, elem):
        new_node = Node(elem)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head = new_node

    def pop(self):
        if self.head is None:
            return None
        else:
            poped_elem = self.head.data
            self.head = self.head.next
            return poped_elem
            
    def top(self):
        if self.head is None:
            return None
        else:
            return self.head.data    

    def is_empty(self):
        return self.head is None

    def print(self):
        if self.head is None:
            print("Stack is empty")
            return
        else:
            print("print stack:")
            curr_node = self.head
            while (curr_node is not None):
                print(curr_node.data)
                curr_node = curr_node.next
            print("print ended", end = ", ")
            return
        
        

```

## Чтение файла и исполнение команд в нём

### Файл input_1

```python
import time

# создание экземпляра класса
mystack_1 = Stack()

file = open("stack_1.txt").read().split()
sup_timer_start = time.perf_counter_ns()

# разбиваем строчку в массив сплитом, проходимся по каждому элементу
for elem in file:
    # мэтч ищется именно по нулевому элементу, чтобы все норм работало в командах по типу "1,cat"(т.е. считывало только 1)
    match elem[0]:
        case '1':
            # пушится elem[2::] потому, что первые два элемента это один и запятая
            print(f"push {elem[2::]}", end=", ")
            time_start = time.perf_counter_ns()
            mystack_1.push(elem[2::])
            res = time.perf_counter_ns() - time_start
            print(f"time: {res / 1_000_000} ms")
        case '2':
            time_start = time.perf_counter_ns()
            poped = mystack_1.pop()
            res = time.perf_counter_ns() - time_start
            print(f"pop {poped}, time: {res /1_000_000} ms")
        case '3':
            time_start = time.perf_counter_ns()
            top = mystack_1.top()
            res = time.perf_counter_ns() - time_start
            print(f"{top} is on top, time: {res / 1_000_000} ms")
        case '4':
            time_start = time.perf_counter_ns()
            is_empty = mystack_1.is_empty()
            res = time.perf_counter_ns() - time_start
            print(f"is stack empty: {mystack_1.is_empty()}, time: {res / 1_000_000} ms")
        case '5':
            time_start = time.perf_counter_ns()
            mystack_1.print()
            res = time.perf_counter_ns() - time_start
            print(f"time: {res / 1_000_000} ms")
        case " ":
            continue

sup_timer_finish = time.perf_counter_ns() 

res = sup_timer_finish - sup_timer_start

print(f"\nвсего на выполнение команд файла ушло {res / 1_000_000} секунд")



# в след примерах все то же самое, так что комментов не будет
```

## Реализация постфиксной формы

```python
stack = Stack()

file = open("postfix.txt").read()
print(f"постфиксная запись: {file}")

res = ""

sup_timer_start = time.perf_counter_ns()
for el in file:
    if el in "+-*:^":
        b, a = stack.pop(), stack.pop()
        res = "(" + a + el + b + ")" if el in "+-" else a + el + b
        stack.push(res)
    else:
        stack.push(el)
        
sup_timer_finish = time.perf_counter_ns() 
timer = sup_timer_finish - sup_timer_start
print(f"инфиксная запись: {res}")
print(f"\nвсего на выполнение команд файла ушло {timer / 1_000_000} секунд")
```

### Алгоритмическая сложность


В общем сложность реализации постфиксной формы с помощью стека равна O(n) из-за прохождения каждого элемента записи. Исходя из того, что операции pop и push занимают o(1), в итоговой оценке они не учавствуют

# Часть 2

```python
class Node:
    def __init__(self, data):
        self.next = None
        self.data = data

    
class Queue:
    def __init__(self):
        self.top = None
        self.tail = None


    def Enqueue(self, elem):
        new_node = Node(elem)
        
        if self.top is None:
            self.top, self.tail = new_node, new_node
            
        elif self.top.next is None:
            self.top.next = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node

    def Dequeue(self):
        try:
            top_elem = self.top.data 
            self.top = self.top.next
            return top_elem
        except AttributeError:
            None

    def first(self):
        if self.top is not None:
            return self.top.data
        else:
            return None

    def is_empty(self):
        return self.top is None
        
    def print(self):
        if self.top is None:
            print("Queue is empty")
        else:
            print("print queue: ")
            curr_node = self.top
            while (curr_node is not None):
                print(curr_node.data)
                curr_node = curr_node.next
            print("print ended", end=", ")
            
```

### 1) Первый вариант (различная длина)

#### Первый инпут

```python
import time

file = open("queue_1.txt")
file_iterator = iter(file)

line = next(file_iterator).split()
queue = Queue()

timers = []


for elem in line:
    match elem[0]:
        case '1':
            print(f"Enqueue {elem[2::]}", end=", ")
            time_start = time.perf_counter_ns()
            queue.Enqueue(elem[2::])
            res = time.perf_counter_ns() - time_start
            timers.append(res / 1_000_000)
            print(f"time: {res / 1_000_000} ms")
        case '2':
            time_start = time.perf_counter_ns()
            poped = queue.Dequeue()
            res = time.perf_counter_ns() - time_start
            timers.append(res / 1_000_000)
            print(f"Dequeue {poped}, time: {res /1_000_000} ms")
        case '3':
            time_start = time.perf_counter_ns()
            top = queue.first()
            res = time.perf_counter_ns() - time_start
            timers.append(res / 1_000_000)
            print(f"{top} is a first elem, time: {res / 1_000_000} ms")
        case '4':
            time_start = time.perf_counter_ns()
            is_empty = queue.is_empty()
            res = time.perf_counter_ns() - time_start
            timers.append(res / 1_000_000)
            print(f"is queue empty: {queue.is_empty()}, time: {res / 1_000_000} ms")
        case '5':
            time_start = time.perf_counter_ns()
            queue.print()
            res = time.perf_counter_ns() - time_start
            timers.append(res / 1_000_000)
            print(f"time: {res / 1_000_000} ms")
        case " ":
            continue


print(f"\nвсего на выполнение команд файла ушло {sum(timers)} миллисекунд")
```
