### подготовка данных

```python
import random
import matplotlib.pyplot as plt
import time
import numpy as np
import pandas as pd
import math
```

```python
def binary_search(arr: list, key: int) -> int:
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        mid_val = arr[mid]
        if mid_val == key:
            return mid
        elif mid_val > key:
            high = mid - 1
        else:
            low = mid + 1
    return -1


def quicksort(arr: list) -> list:
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        less = [i for i in arr[1:] if i < pivot]
        greater = [i for i in arr[1:] if i > pivot]
        equal = [i for i in arr[1:] if i == pivot]

        return quicksort(less) + [pivot] + equal + quicksort(greater)


n_values = [n for n in range(1, 1000, 5)]

sort_time = []
search_time = []
count_time = []
n_graph = []
log_graph = []
```

### Тест сортировки

худший случай быстрой сортировки имеет сложность O(n<sup>2</sup>) и заключается в подаче уже отсортированного массива с первым либо последним опорным элементом, в таком случае на каждом погружении в рекурсию массив, который нужно отсортировать, равняется len(arr) - 1.

```python
for n in n_values:
    arr = [i for i in range(n)]
    timer_start = time.perf_counter_ns()
    quicksort(arr)
    timer_finish = time.perf_counter_ns()
    res = timer_finish - timer_start
    sort_time.append(res / 1_000_000)


parts = np.array_split(np.array(sort_time), 8)

times = np.array([])

for part in parts:
    part = pd.Series(part)
    part = part.where(part < part.quantile(.95))
    times = np.concatenate((times, part))

df = pd.DataFrame({
    "time": times,
    "n": n_values,
})


df = df.dropna()

fig, ax1 = plt.subplots()


ax1.set_xlabel("n")
ax1.set_ylabel("time (ms)")
ax1.plot(df["n"], df["time"])
ax1.tick_params(axis='y')

ax2 = ax1.twinx()
graph = [n**2 for n in n_values]
ax2.plot(n_values, graph, color="navy")

ax1.legend(['График тестов'])
ax2.legend(['График n^2'], loc="lower right")
plt.grid()
plt.show()
    
```

### Тест поиска

бинарный поиск в худшем варианте имеет сложность O(log<sub>2</sub>(n)). Худшим же случаем является, когда нам нужно найти первый либо последний элемент, тогда алгоритм совершает (почти) ровно log<sub>2</sub>(n) делений. На этом графике много скачков, почистить их не удалось(да и чистить нечего, он почти весь состоит из скачков)

```python
for n in n_values:
    arr = sorted([i for i in range(n)])
    key = arr[0]
    timer_start = time.perf_counter_ns()
    binary_search(arr, key)
    timer_finish = time.perf_counter_ns()
    res = timer_finish - timer_start
    search_time.append(res / 1_000_000)



parts = np.array_split(np.array(search_time), 4)

times = np.array([])

for part in parts:
    part = pd.Series(part)
    part = part.where(part < part.quantile(.99))
    times = np.concatenate((times, part))

df = pd.DataFrame({
    "time": times,
    "n": n_values,
})

df = df.dropna()

fig, ax1 = plt.subplots()


ax1.set_xlabel("n")
ax1.set_ylabel("time (ms)")
ax1.plot(df["n"], df["time"])
ax1.tick_params(axis='y')

ax2 = ax1.twinx()
graph = np.log2(n_values)
ax2.plot(n_values, graph, color="navy")
ax2.set_ylim(0, 20)

ax1.legend(['График тестов'])
ax2.legend(['График log(n)'], loc="lower right")
ax1.grid()
plt.show()
```

### тест подсчёта кол-ва значений

count проходится по всему списку подсчитывая значения, поэтому в любом случае имеет сложность O(n)

```python
for n in n_values:
    arr = [1000 for i in range(n)]
    key = arr[0]
    timer_start = time.perf_counter_ns()
    arr.count(key)
    timer_finish = time.perf_counter_ns()
    res = timer_finish - timer_start
    count_time.append(res)




df = pd.DataFrame({
    "time": count_time,
    "n": n_values,
})


df = df.dropna()


fig, ax1 = plt.subplots()


ax1.set_xlabel("n")
ax1.set_ylabel("time (ms)")
ax1.plot(df["n"], df["time"])
ax1.tick_params(axis='y')

ax2 = ax1.twinx()
graph = n_values
ax2.plot(n_values, graph, color="navy")

ax1.legend(['График тестов'])
ax2.legend(['График n'], loc="lower right")
ax1.grid()
plt.show()


```

## Подготовка данных

```python
import random
import matplotlib.pyplot as plt
import time
import math
import numpy as np
import sys

def binary_search(arr: list, key: int) -> int:
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        mid_val = arr[mid]
        if mid_val == key:
            return mid
        elif mid_val > key:
            high = mid - 1
        else:
            low = mid + 1
    return -1


def quicksort(arr: list) -> list:
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        less = [i for i in arr[1:] if i < pivot]
        greater = [i for i in arr[1:] if i > pivot]
        equal = [i for i in arr[1:] if i == pivot]

        return quicksort(less) + [pivot] + equal + quicksort(greater)

sys.setrecursionlimit(10000)

n_values = [5, 50, 5000]
```

## Своя сортировка

Средний и лучший случай в быстрой сортировке имеют почти одинаковую производительность. Если применить отдельную функцию для разделения массива, можно выжать лучшее время работы O(n), но в нашем варианте время работы равна O(n log(n)), как и в среднем случае.

Для худшего случая был выбран уже отсортированный список, для среднего  рандомносгенерированный, для лучшего [ДАННЫЕ_УДАЛЕНЫ]

### Худший случай

```python
for n in n_values:
    arr = [i for i in range(1, n+1)]
    timer_start = time.perf_counter_ns()
    quicksort(arr)
    timer_finish = time.perf_counter_ns()
    res = timer_finish - timer_start
    print(f"{res / 1_000_000} миллисекунд - Время при n = {n}")
```

### Средний случай(и лучший)

```python
for n in n_values:
    arr = [random.randint(1, n) for i in range(n)]
    timer_start = time.perf_counter_ns()
    quicksort(arr)
    timer_finish = time.perf_counter_ns()
    res = timer_finish - timer_start
    print(f"{res / 1_000_000} миллисекунд - Время при n = {n}")
```

## Встроенная сортировка

Python использует сортировку Timsort, в которой худший и средний случай совпадают по времени и имеют оценку O(n * log(n)). 
В лучшем  же случае сложность алгоритма составляет O(n)

Экспериментальным путём мы пришли к тому, что алгоритм лучше всего работает на уже отсортированном массиве(противоположно быстрой сортировке)
Для среднего случая использовался рандомно сгенерированный массив

### Лучший случай

```python
for n in n_values:
    arr = [i for i in range(1, n+1)]
    timer_start = time.perf_counter_ns()
    sorted(arr)
    timer_finish = time.perf_counter_ns()
    res = timer_finish - timer_start
    print(f"{res / 1_000_000} миллисекунд - Время при n = {n}")
```

### Средний(худший) случай

```python
for n in n_values:
    arr = [random.randint(1, n) for i in range(1, n+1)]
    timer_start = time.perf_counter_ns()
    sorted(arr)
    timer_finish = time.perf_counter_ns()
    res = timer_finish - timer_start
    print(f"{res / 1_000_000} миллисекунд - Время при n = {n}")
```

## Свой поиск

Бинарный поиск в среднем и худшем случае имеет оценку O(log(n)), в лучшем же O(1), если искомое значение оказалось в центре. Худший результат
получается при поиске первого либо последнего значения(при условии что значения уникальны). Средний случай - что-то между лучшим и худшим случаем, например когда значение в середине первой или второй половине массива.

### Худший случай

```python
for n in n_values:
    arr = sorted([i for i in range(1, n+1)])
    key = arr[0]
    timer_start = time.perf_counter_ns()
    binary_search(arr, key)
    timer_finish = time.perf_counter_ns()
    res = timer_finish - timer_start
    print(f"{res / 1_000_000} миллисекунд - Время при n = {n}")
```

### Лучший случай

```python
for n in n_values:
    arr = sorted([i for i in range(1, n+1)])
    key = arr[len(arr) // 2]
    timer_start = time.perf_counter_ns()
    binary_search(arr, key)
    timer_finish = time.perf_counter_ns()
    res = timer_finish - timer_start
    print(f"{res / 1_000_000} миллисекунд - Время при n = {n}")
```

### Средний случай

```python
for n in n_values:
    arr = [i for i in range(1, n+1)]
    key = arr[len(arr) // 4]
    timer_start = time.perf_counter_ns()
    binary_search(arr, key)
    timer_finish = time.perf_counter_ns()
    res = timer_finish - timer_start
    print(f"{res / 1_000_000} миллисекунд - Время при n = {n}")
```

## Встроенный поиск

Встроенный поиск имеет худшее и среднее время O(n), лучшее же O(1).
Для худшего выбран последний элемент,
Для среднего элемент посередине,
Для лучшего первый элемент.

### Худший случай

```python
for n in n_values:
    arr = [i for i in range(1, n+1)]
    key = arr[-1]
    timer_start = time.perf_counter_ns()
    arr.index(key)
    timer_finish = time.perf_counter_ns()
    res = timer_finish - timer_start
    print(f"{res / 1_000_000} миллисекунд - Время при n = {n}")
```

### Средний случай

```python
for n in n_values:
    arr = [i for i in range(1, n+1)]
    key = arr[len(arr) // 2]
    timer_start = time.perf_counter_ns()
    arr.index(key)
    timer_finish = time.perf_counter_ns()
    res = timer_finish - timer_start
    print(f"{res / 1_000_000} миллисекунд - Время при n = {n}")
```

### Лучший случай

```python
for n in n_values:
    arr = [i for i in range(1, n+1)]
    key = arr[0]
    timer_start = time.perf_counter_ns()
    arr.index(key)
    timer_finish = time.perf_counter_ns()
    res = timer_finish - timer_start
    print(f"{res / 1_000_000} миллисекунд - Время при n = {n}")
```
